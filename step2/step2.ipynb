```
var model = new Model();

var a = new Vector3();
var b = new Vector3(10, 10);
var c = new Vector3(20, 5);
var d = new Vector3(25, 10);

var points = new List<Vector3>{a, b, c, d};

var pline = new Polyline(points);
var offset = pline.Offset(1, EndType.Square);

model.AddElement(new ModelCurve(pline, BuiltInMaterials.XAxis));
model.AddElement(new ModelCurve(offset[0], BuiltInMaterials.YAxis));

model
```

```
var model = new Model();

var angle_step = 5;
var height = 1;
int n_floor = 10;

for (int i=0; i<n_floor; i++)
{
    var angle = i * angle_step;
    var floor = new Floor(Polygon.Rectangle(10, 10), 0.1, null, BuiltInMaterials.Mass);
    floor.Transform = new Transform(0, 0, i*height);
    floor.Transform.Rotate(Vector3.ZAxis, angle);
    model.AddElement(floor);
}

model
```

```
var model = new Model();

var wood_material = BuiltInMaterials.Wood;
var glass_material = BuiltInMaterials.Glass;

// Floor Profile
var p = Polygon.L(10, 20, 5);
// Floor
var floor = new Floor(p, 0.1, null , wood_material);
// Ceiling
var ceiling = new SolidCeiling(p, 0.1, wood_material, new Transform(0, 0, 3));
// Wall
var wall = new Wall(p, 3.0, glass_material);

// Column
var column_profile = Polygon.Rectangle(0.2, 0.2);
var colums = new List<Column>();
var vertices = p.Vertices;
for (var x = 0; x < vertices.Count; x++)
{
    colums.Add(new Column(vertices[x], 3.0, column_profile, wood_material));
}

model.AddElement(floor);
model.AddElement(ceiling);
model.AddElements(colums);
model.AddElement(wall);

model
```
